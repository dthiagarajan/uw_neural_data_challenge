```python
import pandas as pd
import numpy as np
import os 
import matplotlib.pyplot as plt 

from fastai.tabular import *
from fastai.vision import *

from PIL import Image

import torch
from torch.utils.data import Dataset, DataLoader
import torchvision
from torchvision import transforms

torch.manual_seed(3)
torch.cuda.manual_seed(3)

%matplotlib inline

stim = np.load('../input/stim.npy')
df = pd.read_csv('../input/train.csv')
df.index = df.Id
df = df.fillna(df.mean())
df.Id = ['../data/%d.png' % i for i in range(50, 601)]
```

```python
def get_label(fn):
    idx = int(str(fn).split('/')[-1].split('.')[0]) - 50
    assert idx >= 0
    return df.values[idx].tolist()[0]
```

```python
cols = df.columns.values[1:]
```

```python
import fastai
data = ImageItemList.from_df(df, f'../data').random_split_by_pct() \
                    .label_from_df(cols=list(range(1, 19)), label_cls=fastai.FloatList) \
                    .transform(get_transforms(), size=224).databunch()
data.normalize(imagenet_stats)
```

```python
data.train
```

```python
data.train_dl.num_workers = 0
data.valid_dl.num_workers = 0
learn = create_cnn(data, models.resnet34)

learn.lr_find()
learn.recorder.plot()
```

```python
lr = 1e-1
learn.model(torch.cuda.FloatTensor(stim[0,:]))
```

```python
learn.fit_one_cycle(4, max_lr=lr)
```

```python
learn.save('learner0')
```

```python
learn.unfreeze()
learn.fit_one_cycle(4, max_lr=slice(1e-6, 1e-3))
```

```python
learn.save('learner1')
```
